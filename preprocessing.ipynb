```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the data
file_path = Path("Resources\healthcare-dataset-stroke-data.csv")
df = pd.read_csv(file_path)
df
```

```python
# Evaluate Data: Data types and null values
print(df.dtypes)

print(df.count())
```

```python
# Drop rows if missing value
df = df.dropna(axis=0, inplace=False)
print(df.count())

# Drop Patient ID
df = df.drop(columns=['id'])
```

```python
# Create the output file (CSV).
output_data_file = "resources/processed_stroke_db_w_cat.csv"
# Export the City_Data into a CSV.
df.to_csv(output_data_file)
```

```python
# Change object dtypes to categorical
le = LabelEncoder()
categories = ["gender", "ever_married", "work_type", "Residence_type", "smoking_status"]

for category in categories:
    df[category] = le.fit_transform(df[category]) 


df
```

```python
# Double Check Data: Data types and null values
print(df.dtypes)

print(df.count())
```

```python
# Create the output file (CSV).
output_data_file = "resources/processed_stroke_db.csv"
# Export the City_Data into a CSV.
df.to_csv(output_data_file)
```
